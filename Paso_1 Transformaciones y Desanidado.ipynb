```python
#Importamos las librerias que usaremos
import json
import pandas as pd
import numpy as np
```

```python
#Cargamos el dataset
data = pd.read_csv('movies_dataset.csv')
```

```python
data
```

```python
#Eliminamos las columnas "video","imdb_id","adult","original_title","vote_count","poster_path", "homepage"
data.drop(["video","imdb_id","adult","original_title","vote_count","poster_path", "homepage"], axis=1, inplace=True)
```

```python
#Reemplazamos datos vacios de la columna "revenue" por "0"
data["revenue"]=data["revenue"].fillna(0)
```

```python
#Revisamos cuantos nulos hay en la columna 'release_date'
data['release_date'].isnull().sum()
```

```python
#Eliminamos las filas que contengan 'release_date' nulos
data = data.dropna(subset=['release_date']).reset_index(drop=True)
```

```python
#Revisamos nuevamente cuantos nulos hay en la columna 'release_date'
data['release_date'].isnull().sum()
```

```python
#Convertimos la columna 'release_date' en formato fecha
data.loc[:, "release_date"] = pd.to_datetime(data["release_date"], errors="coerce")
```

```python
#Revisamos nuevamente cuantos nulos hay en la columna 'release_date'
data['release_date'].isnull().sum()
```

```python
#Eliminamos las filas que contengan 'release_date' nulos
data = data.dropna(subset=['release_date']).reset_index(drop=True)
```

```python
#Revisamos nuevamente cuantos nulos hay en la columna 'release_date'
data['release_date'].isnull().sum()
```

```python
#Creamos una nueva columna "release_year" a partir de la columna "release_date"
data["release_year"] = pd.to_datetime(data["release_date"], errors="coerce").dt.year
```

```python
#Revisamos nuevamente cuantos nulos hay en la columna 'release_date'
data['release_year'].isnull().sum()
```

```python
#Convertimos a entero 'release_year'
data['release_year'] = data['release_year'].astype(int)
```

```python
#Reemplazamos datos vacios de la columna "budget" por "0"
data["budget"] = pd.to_numeric(data["budget"], errors="coerce").fillna(0).astype(float)
```

```python
#Creamos una nueva columna "return" a partir de la division entre "revenue" y "budget"
data["return"] = np.where(data["budget"] == 0, 0, data["revenue"] / data["budget"])
```

```python
#Preservamos el dataframe y utilizamos el otro para el desanidado
bc=data
```

```python
#Reemplazamos comillas simples por dobles, asi podemos usar el modulo json en esa columna para  desanidarla
bc['belongs_to_collection'] = bc['belongs_to_collection'].str.replace("'", '"')
```

```python
#Generamos esta funcion para evitar que de error cuando desanidemos la columna aplicamos despues del fillna para rellenar nulos con un diccionario vacio
def safe_json_loads(s):
    try:
        return json.loads(s)
    except:
        return {}
```

```python
#Aplicamos la funcion safe_json_loads despues del fillna para rellenar nulos con un diccionario vacio
bc['belongs_to_collection'] = bc['belongs_to_collection'].fillna('{}')
df_belongs = pd.json_normalize(bc['belongs_to_collection'].apply(safe_json_loads))
```

```python
#Visualizamos el dataframe resultante de desanidar la columna 'belongs_to_collection'
df_belongs
```

```python
#Renombramos las columnas para saber que pertencen a esa 'belongs_to_collection'
df_belongs = df_belongs.rename(columns={"id":"collection_id", "name":"collection_name", "poster_path":"collection_poster_path", "backdrop_path": "collection_backdrop_path"})
```

```python
#Visualizamos el dataframe resultante de desanidar la columna 'belongs_to_collection'
df_belongs
```

```python
#Concatenamos el dataframe resultante de desanidar "belongs_to_collection" al dataframe original y eliminamos la columna belongs_to_collection
data=pd.concat([data, df_belongs], axis=1)
data = data.drop('belongs_to_collection', axis=1)
```
